```python
# Dependencies
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import pymongo
import time 
from IPython import get_ipython
import nbconvert
```

Read in Table from URL as Data Frame

```python
rep_url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States_by_population'

df = pd.read_html(rep_url)[0]
df
```

Flatten Multi-Level Index

```python
df.columns = df.columns.get_level_values(0)

#df.columns = [' '.join(col).strip() for col in df.columns.values]

df
```

Select Relevant Columns

```python
df = df[['State', 'Census population', 'Total U.S. House of Representatives Seats']]
df
```

Rename 2010 Census Column to Address Duplicate Column Names

```python
column_names = df.columns.values
column_names[2] = "2010 Census Population"
df.columns = column_names
df
```

Drop Estimated Unnecessary Estimated Population Column

```python
df = df.drop(df.columns[1], axis=1)
df
```

Replace Values for House of Representatives Seats w/ NaN for non-State Rows and Drop NaNs

```python
representation_df = df.replace({'1 (non-voting)': np.NaN, '435 (+ 1 non-voting)': np.NaN, '435 (+ 6 non-voting)': np.NaN, 
                         '432': np.NaN, '435': np.NaN})

representation_df = representation_df.dropna()
representation_df
```

Reset Index to Avoid Confusion

```python
representation_df = representation_df.reset_index(drop=True)

representation_df
```

```python
representation_df['Senate Seats'] = 2
electoral_college_col = representation_df['Total U.S. House of Representatives Seats'].astype(int) + 2
representation_df['Electoral College Votes'] = electoral_college_col
representation_df = representation_df.rename(columns = {'Total U.S. House of Representatives Seats': 'House of Reps Seats'})
representation_df
```
