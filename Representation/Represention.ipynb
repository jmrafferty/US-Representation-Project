```python
# Dependencies
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import pymongo
import time 
from IPython import get_ipython
import nbconvert
```

Read in Table from URL as Data Frame

```python
rep_url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States_by_population'

df = pd.read_html(rep_url)[0]
df
```

Flatten Multi-Level Index

```python
df.columns = df.columns.get_level_values(0)

#df.columns = [' '.join(col).strip() for col in df.columns.values]

df
```

Select Relevant Columns

```python
df = df[['State', 'Census population', 'Total U.S. House of Representatives Seats']]
df
```

Rename 2010 Census Column to Address Duplicate Column Names

```python
column_names = df.columns.values
column_names[2] = "2010 Census Population"
df.columns = column_names
df
```

Drop Estimated Unnecessary Estimated Population Column

```python
df = df.drop(df.columns[1], axis=1)
df
```

Replace Values for House of Representatives Seats w/ NaN for non-State Rows and Drop NaNs

```python
representation_df = df.replace({'1 (non-voting)': np.NaN, '435 (+ 1 non-voting)': np.NaN, '435 (+ 6 non-voting)': np.NaN, 
                         '432': np.NaN, '435': np.NaN})

representation_df = representation_df.dropna()
representation_df
```

Reset Index to Avoid Confusion

```python
representation_df = representation_df.reset_index(drop=True)

representation_df
```

```python
representation_df['senate_seats'] = 2
electoral_college_col = representation_df['Total U.S. House of Representatives Seats'].astype(int) + 2
representation_df['electoral_college_votes'] = electoral_college_col
representation_df = representation_df.rename(columns = {'2010 Census Population': '2010_census_population',
                                                        'Total U.S. House of Representatives Seats': 'house_of_reps_seats'
                                                       })
representation_df['pop_per_senator_(mil)'] = representation_df['2010_census_population'].astype(int) / representation_df['senate_seats'] / 1000000
representation_df['pop_per_rep_(mil)'] = representation_df['2010_census_population'].astype(int) / representation_df['house_of_reps_seats'].astype(int) / 1000000
representation_df
```

Read in Federal Funding CSV (Soure: https://worldpopulationreview.com/state-rankings/federal-aid-by-state)

```python
funding_df = pd.read_csv('federal-aid-by-state.csv')
funding_df
```

```python
funding_df = funding_df.rename(columns = {'State':'state','federalFunding': 'per_capita_fed_funding_($)', 
                                         'totalFederalFunding': 'total_fed_funding_(billions_of_$)', 
                                         'benefitsRecipients': 'benefits_receipients_(%)',
                                        'medianHouseholdIncome': 'Median Household Income ($)'})

funding_df['total_fed_funding_(billions_of_$)'] = funding_df['total_fed_funding_(billions_of_$)'] / 1000000000

funding_df
```

```python
elect_url = 'https://en.wikipedia.org/wiki/List_of_United_States_Senate_election_results_by_state'

elect_df = pd.read_html(elect_url)[2]

elect_df
```

```python
elect_df = elect_df.head(50)
#elect_df.drop(elect_df.index([[50, 57]])
#elect_df.index
elect_df
```

```python
elect_df = elect_df.dropna(axis=1, inplace = False)
elect_df
```

```python
elect_df = elect_df.iloc[:, 1:]
elect_df.columns = elect_df.columns.get_level_values(1)

column_names2 = elect_df.columns.values
column_names2[0] = "state"
column_names2[1] = "dem_votes_1932-66"
column_names2[2] = "rep_votes_1932-66"
column_names2[3] = "other_votes_1932-66"
column_names2[4] = "%_dem_1932-66"
column_names2[5] = "dem_votes_1968-2016"
column_names2[6] = "rep_votes_1968-2016"
column_names2[7] = "other_votes_1968-2016"
column_names2[8] = "%_dem_votes_1968-2016"

elect_df.columns = column_names2
elect_df
```

```python
elect_df = elect_df[["state", "dem_votes_1968-2016", "rep_votes_1968-2016", 
                    "other_votes_1968-2016", "%_dem_votes_1968-2016"]]

elect_df
```
